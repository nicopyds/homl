```python
from zlib import crc32

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./datasets/housing.csv")
```

```python
df.reset_index(inplace=True)
```

```python
ids = df["index"]
is_id_in_test = ids.apply(lambda id_: crc32(np.int64(id_)) < (0.2 * 2**32))
```

```python
is_id_in_test.sum()
```

```python
df["income_cat"] = pd.cut(
    x=df["median_income"],
    bins=[0.0, 1.5, 3.0, 4.5, 6.0, np.inf],
    labels=[1, 2, 3, 4, 5],
)
```

```python
df["income_cat"].value_counts().sort_index().plot(kind="bar")
```
